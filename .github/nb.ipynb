Навчився створювати віртуальні середовище Пайтон
Створення віртуального середовища:
python -m venv ./my_env
Ця команда створює нове віртуальне середовище з назвою my_env в поточній директорії.

Активація віртуального середовища:
`source my_env/Scripts/activate`
Ця команда активує віртуальне середовище, що дозволяє вам ізолювати установлені пакети та середовище Python.

Встановлення бібліотеки requests:
`pip install requests`
Ця команда встановлює бібліотеку requests в активоване віртуальне середовище.

Деактивація віртуального середовища:
`deactivate`
Ця команда деактивує віртуальне середовище.

Перевірка встановленої бібліотеки:
`pip show requests`
Ця команда показує інформацію про встановлену бібліотеку requests, таку як версія, автори, залежності тощо.

Ці команди допоможуть вам створити віртуальне середовище, встановити в нього бібліотеку та перевірити результати.
`pipenv` - це інструмент управління залежностями та віртуальними середовищами для Python, який поєднує функціональність `pip` та `virtualenv`. Він надає зручний інтерфейс для роботи з проектами Python. Ось деякі команди, які можна виконувати за допомогою `pipenv`:

1. Ініціалізація проекту `pipenv`:

```bash
pipenv --python 3.8
```

Ця команда створює новий проект `pipenv` з використанням Python 3.8.

2. Встановлення пакетів:

```bash
pipenv install <package_name>
```

Ця команда встановлює пакети в проект `pipenv`.

3. Встановлення пакетів з файлу `Pipfile`:

```bash
pipenv install
```

Ця команда встановлює всі пакети, які перераховані в файлі `Pipfile`.

4. Запуск команд у віртуальному середовищі:

```bash
pipenv run <command>
```

Ця команда виконує команду в контексті віртуального середовища `pipenv`.

5. Активація віртуального середовища:

```bash
pipenv shell
```

Ця команда активує віртуальне середовище `pipenv`.

6. Перевірка залежностей:

```bash
pipenv graph
```

Ця команда відображає граф залежностей проекту.

7. Оновлення пакетів:

```bash
pipenv update
```

Ця команда оновлює всі пакети до їх останніх версій, відповідно до вказаних версій у файлі `Pipfile`.
